This file will be for cleaning the Kaggle dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FPL/data/players.csv')
df.head()
```

```python
del_columns = set(df.columns) - set(['name', 'now_cost', 'position', 'team', 'total_points'])

df.drop(del_columns, axis=1, inplace=True)
df.head()
```

```python
df.rename(columns={'now_cost': 'cost', 'total_points': 'points'}, inplace=True)
df.head()
```

```python
df_encoded = pd.get_dummies(df, columns=['position', 'team'], dtype=int)
df_encoded.head()
```

```python
df_encoded.to_csv('FPL/data/players_cleaned.csv', index=False)
```
